```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(20)
y = np.arange(40, 60)
```

```python
plt.scatter(x, y, c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('aaaa')
plt.plot(x, y, c='r')
plt.subplot(2,2,2)
plt.xlabel('x1 axis')
plt.ylabel('y1 axis')
plt.title('bbbb')
plt.plot(x, y, c='b')
plt.show()
```

```python
#Bar Plot
x = [2, 4, 6]
y = [4, 6, 8]
x1 = [3,9, 11]
y2 = [10, 20, 30]
plt.bar(x, y, color='r')
plt.bar(x1, y2, color='g')
plt.figure(figsize=(10,6))
plt.show()
```

```python
import random

num_integers = 40

random_integers = [random.randint(1, 40) for _ in range(num_integers)]

# print(random_integers)
```

```python
#Histogram

x= np.array(random_integers)
plt.hist(x)
plt.title("histogram")
plt.show()
```

```python
#boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```
