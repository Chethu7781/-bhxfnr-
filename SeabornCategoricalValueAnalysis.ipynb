##CATEGORICAL PLOTS
Seaborn also helps us in doing analysis on categorical data points.


*   Boxplot
*   Voilinplot
*   countplot
*   bar plot




```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
sns.countplot(data=df, x='sex')
```

```python
sns.barplot(data=df, x='sex', y='total_bill')
```

```python
sns.boxplot(x='total_bill', y='sex', data=df)
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

###Voilinplot

Voilinplot helps us to see both the distribution of the data in terms of Kernel density estimation and the boxplot

```python
sns.violinplot(x='total_bill', y='sex', data=df, hue='smoker')
```
